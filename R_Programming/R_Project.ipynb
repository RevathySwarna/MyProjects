#### Installing the needed packages

```r
install.packages("tidyverse")
```

#### Importing libraries

```r
library(rvest)
library(tidyr)
library(dplyr)
library(ggplot2)
library(tidyverse)
library(tibble)
library(lattice)
```

#### Reading the url

```r
url <- "https://en.wikipedia.org/wiki/World_Happiness_Report"
page <- read_html(url)
```

#### Scraping the table content

```r
table <- page %>%
  html_nodes("table.wikitable") %>%
  .[[3]]
```

```r
table_data <- table %>%
  html_table(fill = TRUE)
```

```r
table_data
```

#### Saving the content as csv file

```r
write.csv(table_data, "table_data.csv", row.names = FALSE)
```

#### Reading the csv file

```r
data <- read.csv("table_data.csv")
```

```r
data
```
